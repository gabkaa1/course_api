```python
import asyncio
import time
```

```python
async def get_large_data():
    print('start - large')
    await asyncio.sleep(3)
    print('end - large')

async def get_small_data():
    print('start - small')
    await asyncio.sleep(1)
    print('end - small')

async def main1():
    start = time.perf_counter()

    await get_large_data()
    await get_small_data()

    end = time.perf_counter()
    print(f'lasts for: {end-start:.2f}') # спецификатор :-начало спецификации, .2-округление до 2 знаков после запятой, f-формат float
    # {value: [заполнитель][выравнивание][ширина][.точность]f}

async def main2():
    start = time.perf_counter()

    task1 = asyncio.create_task(get_large_data())
    task2 = asyncio.create_task(get_small_data())
    await task1
    await task2

    end = time.perf_counter()
    print(f'lasts for: {end-start:.2f}')
```

```python
await main1()
```

```python
await main2()
```
